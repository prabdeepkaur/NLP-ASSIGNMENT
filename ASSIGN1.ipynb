```python
text = """Artificial Intelligence is transforming the world at a rapid pace.
AI helps in healthcare by predicting diseases, assisting doctors, and managing patient data.
In education, AI provides personalized learning experiences, intelligent tutoring systems, and smart content.
Finance industries use AI for fraud detection, stock market prediction, and risk management.
Machine learning and deep learning are important branches of AI that allow systems to learn from data and improve over time.
With AI, machines can analyze vast amounts of information, recognize complex patterns, and make autonomous decisions.
The future of technology depends heavily on artificial intelligence, automation, and robotics.
While AI brings many opportunities, it also raises ethical concerns, job displacement issues, and the need for responsible innovation.
Researchers and policymakers are working together to ensure AI benefits society in a safe and fair manner."""
```

```python
#1. Tokenize into words.
words = nltk.word_tokenize(text.lower())
```

```python
#2. Remove stopwords.
stop_words = set(stopwords.words("english"))
words = [w for w in words if w.isalpha() and w not in stop_words]
print("✅ Cleaned Tokens:", words[:30])
```

```python
#3. Create a word cloud.
wc = WordCloud(width=600, height=350, background_color="white").generate(" ".join(words))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#4. Generate and print top 10 bigrams and trigrams with frequencies.
bigrams = list(ngrams(words, 2))
trigrams = list(ngrams(words, 3))
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)
```

```python
print("\n🔹 Top 10 Bigrams:")
for bigram, freq in bigram_freq.most_common(10):
    print(bigram, ":", freq)

print("\n🔹 Top 10 Trigrams:")
for trigram, freq in trigram_freq.most_common(10):
    print(trigram, ":", freq)
```
